Ej. 2: Ingresar las notas de los dos parciales de un alumno e indicar si promocionó, aprobó o debe recuperar.
- Se promociona cuando las notas de ambos parciales son mayores o iguales a 7.
- Se aprueba cuando las notas de ambos parciales son mayores o iguales a 4.
- Se debe recuperar cuando al menos una de las dos notas es menor a 4.

*Pista: Usar estructura condicional*

```
# Sin usar if anidados
nota1 = int(input('Cual es la nota del primer parcial?')) 
nota2 = int(input('Cual es la nota del segundo parcial?'))

if (nota1 >= 7 and nota2 >= 7):
  print('Promocionaste la materia')
 
if (nota1 >= 4 and nota2 >= 4): 
  print('Aprobaste la materia')

if (nota1 <4 or nota1 <4):
  print('Debe recuperar')
```

```
# Usando if anidados
nota1 = int(input('Cual es la nota del primer parcial?')) 
nota2 = int(input('Cual es la nota del segundo parcial?'))

if (nota1 >= 7 and nota2 >= 7):
  print('Promocionaste la materia')
else:
  if (nota1 >= 4 and nota2 >= 4): 
    print('Aprobaste la materia')
  else:
    print('Debe recuperar')
```

Ej. 3:	Se desea ingresar la nota de 15 alumnos y mostrar su promedio.

*Pista: Usar estructura iterativa ¿Cuál de las dos que vimos?*

```
# promedio = suma_notas / cantidad

suma_notas = 0 #Esta variable es un "acumulador" porque va sumando numeros
cantidad = 0   #Esta variable es un "contador" porque se incrementa de a 1

for i in range (1,16):
  nota = int(input('Ingresar nota del alumno:'))
  suma_notas = suma_notas + nota
  cantidad = cantidad + 1

promedio = suma_notas / cantidad
print(f'El promedio es: {promedio}')
```

Ej. 4: Ingresar números por teclado y contar la cantidad de números positivos. La carga de datos finaliza cuando se ingresa un 0. Al finalizar, mostrar la cantidad de positivos que se ingresó.

*Pista: Usar estructura iterativa ¿Cuál de las dos que vimos?*


```
# Numeros enteros: 1, 2, 3, 4, 5, -1, -2, -3, 0
# Numeros reales: 1.2, 1.2, 1.241255, -5.6
# Numeros naturales: 1, 2, 3, 4

cant_positivos = 0 # Contador

numero = float(input('Ingrese un numero (0 para finalizar): ')) #1. Inicializo la variable de control
while (numero != 0): #2. Evaluo la variable de control
  if (numero > 0):
    cant_positivos = cant_positivos + 1
  numero = float(input('Ingrese un numero (0 para finalizar): ')) #3. Actualizar el valor de la variable de control para que no sea un bucle infinito

print(f'Se ingresaron {cant_positivos} números positivos')
```

Ej. 5: Leer dos números naturales A y B. Calcular A elevado a la B mediante productos sucesivos y mostrar el resultado.

```
a = int(input('Ingrese A: '))
b = int(input('Ingrese B: '))

resultado = a # esta variable es un acumulador

for i in range(1, b):
  resultado = resultado * a

print(f'{a}^{b}={resultado}')

  
```

# Funciones

En matemática: por ejemplo y(x) = x * x

- Hay una variable "x" que es lo que entra a la función.
- La función ejecuta alguna acción con esa variable.
- Y después le asigna el resultado de la función a "y".

Retomemos el tema de funciones. Como dijimos en la primer clase de repaso, las funciones sirven principalmente para:

- Resolver operaciones matemáticas
- Son bloques de código que se pueden reutilizar
- Si un bloque de código se repite varias veces, en lugar de reescribirlo conviene escribir una función.
- Cada función resuelve una tarea específica y no muchas cosas (en ese caso habría que escribir más de una función).

En python diferenciamos dos tipos de funciones: 
- Built-in functions (funciones que ya vienen integradas con python)

Ya hemos utilizado previamente los dos tipos de funciones, pero quizá sin conocer esta diferenciación.

Ahora pasemos a los detalles de las **funciones definidas por el usuario**:

En python para definir nuestras propias funciones tenemos que respetar la siguiente sintaxis:


```
def NOMBRE(LISTA_DE_PARAMETROS):
    """DOCSTRING_DE_FUNCION""" <OPCIONAL>
    SENTENCIAS <NO OPCIONAL>
    RETURN [EXPRESION] <OPCIONAL, ES BUENA PRÁCTICA QUE ESTÉ>
```
- NOMBRE, es el nombre de la función.
- LISTA_DE_PARAMETROS, es la lista de parámetros que puede recibir una función.
- DOCSTRING_DE_FUNCION, es la cadena de caracteres usada para documentar la función.
- SENTENCIAS, es el bloque de sentencias en código fuente Python que realizar cierta operación dada.
- RETURN, es la sentencia return en código Python.
- EXPRESION, es la expresión o variable que devuelve la sentencia return.

# 1. Ordenamiento avanzado de listas

Un tema importante cuando de manejo de listas se trata, es poder ordenar las listas según diferentes criterios.

Es ahí donde entran en juego los "algoritmos de ordenamiento".

Veremos una implementación del algoritmo llamado "Bubble sort" que es muy utilizado.

````python
# SWAP - Intercambiar los valores de 2 variables
var_a = 2
var_b = 5

aux = var_a
var_a = var_b
var_b = aux

print(f'a: {var_a}')
print(f'b: {var_b}')
````

````python
# ordenar la lista de menor a mayor utilizando bubble sort
lista = [3,4,1,2]
lista2 = [32,4,123,2.5]

def ordenar(lista):
  for i in range(0,len(lista) - 1):
    for j in range(i+1, len(lista)):
      if (lista[j] < lista[i]):
        #Intercambiar los valores -SWAP
        aux = lista[j]
        lista[j] = lista[i]
        lista[i] = aux
  return lista

ordenada1 = ordenar(lista)
print(ordenada1)

ordenada2 = ordenar(lista2)
print(ordenada2)
````

Ej.2: Se cuenta con dos listas apareadas, una de nombres y una de IDs.

Se pide ordenar las listas por ID.

````python
def ordenar_2_listas(lista_nombres, lista_ids):
  for i in range(0,len(lista_ids) - 1):
    for j in range(i+1, len(lista_ids)):
      if (lista_ids[j] < lista_ids[i]):
        #Intercambiar los valores -SWAP
        aux = lista_ids[j]
        lista_ids[j] = lista_ids[i]
        lista_ids[i] = aux

        #Cada vez que hago un swap, hago el mismo swap sobre la otra lista
        aux = lista_nombres[j]
        lista_nombres[j] = lista_nombres[i]
        lista_nombres[i] = aux

  return lista_nombres, lista_ids

nombres = ['Belen', 'Jorge', 'José', 'Manuel', 'Fernanda']
ids = [1, 4, 3, 5, 2]

print('Listas originales:')
print(nombres)
print(ids)

print('Listas ordenadas:')
lista1, lista2 = ordenar_2_listas(nombres, ids)
print(lista1)
print(lista2)
````

Declaremos una matriz de 3x3

````python
matriz = []
for i in range(0,3):
    matriz.append([])
    for j in range(0,3):
        matriz[i].append(i)
````

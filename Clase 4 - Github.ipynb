# Repositorios

**¿Qué es un repositorio?**

Un repositorio es un espacio centralizado donde se almacena, organiza, mantiene y difunde información digital, habitualmente archivos informáticos, que pueden contener trabajos científicos, conjuntos de datos o software.


---


**¿Por qué es importante trabajar con repositorios cuando se está trabajando de manera colaborativa entre varias personas?**


- Los demás pueden acceder a tus últimos cambios
- Uno puede acceder a los últimos cambios de los demás
- Sirve de backup (no tener localmente los archivos importantes sino en la nube)
- Es una forma ordenada y clara de disponibilizar información
- Un repositorio ordenado ya es "documentación"

# Pasos para crear un repositorio

Entrar a https://github.com/

Create repository e ingresamos:
- nombre del repositorio
- accesibilidad (público, privado)
- licencias (en nuestro caso lo dejaremos en blanco por ahora)- .gitignore (en nuestro caso lo dejaremos en blanco)
- readme (completarlo con la informacion del grupo)
Y le damos clic a CREAR

Para clonar un repositorio de github:

1. Conseguir la URL del repositorio: https://github.com/manquintana/taller_programacion_2_uade
2. Pararse en la carpeta local (en su computadora) donde quieran hacer una copia y abrir la terminal.
3. git clone url_del_repositorio (Para crear la copia local)
4. Hacen las modificaciones que tengan que hacer
5. git add . (Para elegir que archivos se van a comittear, el "." es un comodin)

   git commit -m "Mensaje con los cambios hechos" 
6. git push origin master

Si no queremos escribir en la consolita y preferimos hacer clics: https://tortoisegit.org/
